```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the notch filter
fs = 2.0  # Sampling frequency (Hz)
f0 = 1.0  # Frequency to be removed from signal (Hz)
bw = 0.5
w0 = 2*np.pi*f0
Q = w0/bw   # Quality factor

# Frequency array
freq = np.linspace(0, fs, 1000)  # Frequency array for plotting

# Compute the frequency response of the notch filter
w = 2*np.pi*freq
H = np.abs(np.emath.sqrt(1 - (2*Q**2 * w**2) / ((w0**2 - w**2)**2 + w**2 * w0**2/Q**2)))
H_db = 20 * np.log10(H)  # Convert to dB

# Plot
plt.figure()
plt.plot(freq, H_db, 'b')
plt.title('Frequency Response of the Notch Filter')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude [dB]')
plt.grid(True)
plt.show()
```

```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
fs = 2  # Sample frequency (Hz)
f0 = 1.0  # Frequency to be removed from signal (Hz)
bw = 1.0
w0 = 2*np.pi*f0
Q = w0/bw   # Quality factor
# Design notch filter
b, a = signal.iirnotch(f0, Q, fs*2)
```

```python
# freq
```

```python
# Frequency response
freq, h = signal.freqz(b, a, fs=fs)
# Plot
fig, ax = plt.subplots(2, 1, figsize=(8, 6))
ax[0].plot(freq, 20*np.log10(np.abs(h)), color='blue')
ax[0].set_title("Frequency Response")
ax[0].set_ylabel("Amplitude (dB)", color='blue')
# ax[0].set_xlim([0, 100])
# ax[0].set_ylim([-25, 10])
ax[0].grid(True)
ax[1].plot(freq, np.unwrap(np.angle(h))*180/np.pi, color='green')
ax[1].set_ylabel("Angle (degrees)", color='green')
ax[1].set_xlabel("Frequency (Hz)")
# ax[1].set_xlim([0, 100])
# ax[1].set_yticks([-90, -60, -30, 0, 30, 60, 90])
# ax[1].set_ylim([-90, 90])
ax[1].grid(True)
plt.show()
```

```python
# Parameters for the notch filter
fs = 2.0  # Sampling frequency (Hz)
f0 = 0.5  # Frequency to be removed from signal (Hz)
# a = 0.6  # Controls the bandwidth of the notch

for a in [0.5,0.6,0.7]:
    bw = 2*np.log(a)
    print(a)
    # Frequency array
    freq = np.linspace(0, fs/2, 1000)  # Frequency array for plotting

    # Compute the frequency response of the notch filter
    w = 2*np.pi*freq/fs
    w0 = 2*np.pi*f0/fs

    z = np.exp(1j*w)
    H = np.abs((1 - 2*np.cos(w0)*z**-1 + z**-2) / (1 - 2*a*np.cos(w0)*z**-1 + a**2*z**-2))
    norm = (1-2*np.cos(w0)+1)/(1-2*np.cos(w0)*a+a**2)
    H_db = H-np.max(H)#/norm  # Convert to dB

    # Plot
    plt.plot(freq, H_db)
plt.title('Frequency Response of the Notch Filter')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude [dB]')
plt.grid(True)
```
